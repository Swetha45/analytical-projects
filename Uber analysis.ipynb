# to convert Date/time string into the pandas object use pandas.to_datetime method

```python
data["Date/Time"]=data["Date/Time"].map(pd.to_datetime)
data.tail()
```

# to add one feild to the above dataframe create function

```python
def dofmonth(mn):
    return mn.month
data["Month"]=data["Date/Time"].map(dofmonth)
data.tail()
```

# to retrive day from the datetime


```python
def dayof(dy):
    return dy.day
data["Day"]=data["Date/Time"].map(dayof)
data.head()
```

# to retrive year and weekday, hour from the datetime

```python

def wk(w):
    return w.weekday()
data["Weekday"]=data["Date/Time"].map(wk)
data.head()


def hr(h):
    return h.hour
data["Hour"]=data["Date/Time"].map(hr)
data.head()
```

# to drop the cells

```python
data.drop(['weekday', 'hour'], axis=1)
data.drop(['weekday', 'dom'], axis=1)
```

#cross analysis

```python
def getwh(we):
    return len(we)
by_cross = data.groupby('Weekday Hour'. split()).apply(getwh).unstack()

by_cross.head()
sb.heatmap(by_cross, color='r')
```
